```python
 # Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data= ("C:/Users/Anum Ali/Desktop/GT_Coding_Bootcamp/HOMEWORK/Pandas/schools_complete.csv")
student_data = ("C:/Users/Anum Ali/Desktop/GT_Coding_Bootcamp/HOMEWORK/Pandas/students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#DISTRICT SUMMARY: Create a high level snapshot (in table form) of the district's key metrics
#Calculate the total number of schools
total_schools = len(school_data_complete["School ID"].unique())


#Calculate the total number of students
total_students = len(school_data_complete["Student ID"])


#Calculate the total budget $$$
#First create a new table that only includes each school name once
budget_per_school = school_data_complete.groupby(["school_name"])
#The budget for each school will be equal to the mean budget in the groupby (average of a bunch of equivalent values = value)
mean_table = budget_per_school.mean()
#Add up the budgets to get total budget 
total_budget = mean_table["budget"].sum()
total_budget

#Calculate average math score
avg_math = school_data_complete["math_score"].mean()
#Calculate average reading score
avg_read = school_data_complete["reading_score"].mean()

#Precentage passing math (>=70)
pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
percent_pass_math = len(pass_math)/total_students *100

#Precentage passing reading (>=70)
pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
percent_pass_read = len(pass_read)/total_students *100

#Calculate overall passing rate (fyi--I don't think this is the best way to do it, but this is the way it said to do it on the instructions)
percent_pass_overall = (78.985371 + 81.87784)/2


#Create a dataframe to hold above results
district_summary = {"Total Schools": [total_schools], 
                   "Total Students": [total_students],
                   "Total Budget": [total_budget],
                   "Average Math Score": [avg_math],
                   "Average Reading Score" : [avg_read],
                    "% Passing Math" : [percent_pass_math],
                    "% Passing Reading" : [percent_pass_read],
                   "% Overall Passing Rate" : [percent_pass_overall]
                   }

district_summary_df = pd.DataFrame(district_summary)


#Optional formatting for numbers in District Summary table
district_summary_df.style.format({
    'Total Students': '{:,.0f}'.format,
    'Total Budget': '${:,.2f}'.format,
    'Average Math Score' : '{:,.6f}'.format,
    'Average Reading Score' : '{:,.6f}'.format,
    '% Passing Math' : '{:,.6f}%'.format,
    '% Passing Reading' : '{:,.6f}%'.format,
    '% Overall Passing Rate' : '{:,.6f}%'.format
}
)

```

```python
#SCHOOL SUMMARY: Create an overview table that summarizes key metrics about each school
#Create a dataframe that groups schools by name
grouped_school_df = school_data_complete.groupby(["school_name"])

#School Type
school_type_S = grouped_school_df["type"].first()

#Total Students per school
total_students_S = grouped_school_df["Student ID"].count()

#Total School Budget per school
total_budget_S = grouped_school_df["budget"].first()

#Per Student Budget
per_student_S = total_budget_S/total_students_S

#Average Math Score
avg_math_S = grouped_school_df["math_score"].mean()

#Average Reading Score
avg_reading_S = grouped_school_df["reading_score"].mean()

#% Passing Math
pass_math_S = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()
percent_pass_math_S = pass_math_S/total_students_S*100

#% Passing Reading
pass_read_S = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()
percent_pass_read_S = pass_read_S/total_students_S*100

#Overall Passing Rate
overall_pass_S = (percent_pass_read_S + percent_pass_math_S)/2


school_summary_table = pd.DataFrame({"School Type" : school_type_S,
                                    "Total Students" : total_students_S,
                                    "Total School Budget" : total_budget_S,
                                    "Per Student Budget" : per_student_S,
                                    "Average Math Score" : avg_math_S,
                                    "Average Reading Score" : avg_reading_S,
                                     "% Passing Math" : percent_pass_math_S,
                                    "% Passing Reading" : percent_pass_read_S,
                                     "% Overall Passing Rate": overall_pass_S}
                                   )

#Optional formatting for numbers in School Summary table
school_summary_table.style.format({
    "Total Students": "{:,.0f}".format,
    "Total School Budget": "${:,.2f}".format,
    "Per Student Budget" : "${:,.2f}".format,
    "% Passing Math" : "{:,.6f}%".format,
    "% Passing Reading" : "{:,.6f}%".format,
    "% Overall Passing Rate" : "{:,.6f}%".format
}
)
```

```python
#5TOP PERFORMING SCHOOLS 
top_five_sort = school_summary_table.sort_values("% Overall Passing Rate", ascending = False)
top_five_sort.head(5)
```

```python
#5BOTTOM PERFORMING SCHOOLS
bottom_five_sort = school_summary_table.sort_values("% Overall Passing Rate", ascending = True)
bottom_five_sort.head(5)
```
